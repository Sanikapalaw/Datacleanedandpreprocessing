```python
for col in cmc.select_dtypes(include='object').columns:
    cmc[col] = cmc[col].astype(str).str.strip().str.lower()
```

```python
cmc.columns
```

```python
median_income = cmc['Income'].median()
cmc['Income'].fillna(median_income, inplace=True)
```

```python
cmc.isnull().sum()
```

```python
cmc.columns = cmc.columns.str.lower().str.replace(' ', '_')
```

```python
cmc['year_birth'] = cmc['year_birth'].astype(int)
```

```python
cmc.to_csv('cleaned_data.csv', index=False)
```
